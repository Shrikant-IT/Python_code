```python
#violating the SRP principle....generating and printing


class Report:
    def __init__ (self, data):
        self.data = data
        
     def generate_report(self):
         return self.data

     def print_report(self):
         return (self.generate_report())
```

```python
# SRP Principle -> Separate responsibilities
# One class generates report
# Another class prints report

class Report:
    def __init__(self, data):
        self.data = data

    def generate_report(self):
        return self.data


class ReportPrinting:
    def __init__(self, report_data):
        self.report_data = report_data

    def print_report(self):
        return self.report_data.generate_report()


# Usage
report = Report("Sales Report Data")
printer = ReportPrinting(report)

print(printer.print_report())
```

```python
# 2)Open/Close Principle
-a class should be open for extension but closed for Modification/Change/later/Edit
```

```python
#violating the Open/Close principle....if we get new requiremt for discount_type


class Discount:
    def apply_discount(self, price, discount_type):
        if discount_type=="percentage":
            return price*0.9
        elif discount_type=="fixed":
            return price-10
        else:
            return price
```

```python
d = Discount()

print(d.apply_discount(1000, "percentage"))
print(d.apply_discount(1000, "fixed"))
```

```python
# Correct way:

class Discount: #parent class
    def apply_discount(self, price):
        return price

class PercentageDiscount(Discount): #child class..inheritence of above parent class-Discount
    def apply_discount(self, price):
        return price* 0.9

class FixedDiscount(Discount): #inheritence of above parent class-Discount
    def apply_discount(self, price):
        return price-10

class BBDDiscount(Discount): #inheritence of above parent class-Discount
    def apply_discount(self, price):
        return price*2
```

```python
# Object creation
p = PercentageDiscount()
f = FixedDiscount()
b = BBDDiscount()

# Method calling
print(p.apply_discount(1000))
print(f.apply_discount(1000))
print(b.apply_discount(1000))
```

```python
class shape:
    def area(self):
        pass

class Rectangle(shape):
    def __init__ (self, width, height):
        self.width = width
        self.height = height

    def area(self):
        self.width*self.height

class Square(Rectangle):
    def __init__(self, side):
        self.side=side

    def area(self):
        return self.side*self.side
```

```python
##violating 

#code to find sum of even numbers

def sum_of_even_numbers(numbers):
    total_sum = 0
    for i in range(len(numbers)):
        if numbers[i]%2==0:
            total_sum = total_sum+numbers[i]
    return total_sum   

numbers=[2,3,11,3,66,332,10]
sum_of_even_numbers(numbers)    
```

```python
#Naming Convention..pep8 standards.

Type	Naming Convention	Examples

Function	Use a lowercase word or words. Separate words by underscores to improve readability.
eg:function, python_function

Variable	Use a lowercase single letter, word, or words. Separate words with underscores to improve readability.
eg:x, var, python_variable

Class	Start each word with a capital letter. Don’t separate words with underscores. 
This style is called camel case or Pascal case.
eg:Model, PythonClass

Method	Use a lowercase word or words. Separate words with underscores to improve readability.
eg:class_method, method

Constant	Use an uppercase single letter, word, or words. Separate words with underscores 
to improve readability.
eg:CONSTANT, PYTHON_CONSTANT, PYTHON_LONG_CONSTANT

Module	Use a short, lowercase word or words. Separate words with underscores to improve readability.
eg:module.py, python_module.py


Package	Use a short, lowercase word or words. Don’t separate words with underscores.
eg:package, pythonpackage


```
