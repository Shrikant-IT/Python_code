```python
# ask number from user
# get diff of that number from 20
# if result is positive then return double the diff
# if result is negative then return diff
# if same then return message that "entered number is same"
```

```python
def num_check(n):
    diff = 20-n
    if diff>0:
        # print(diff*2)
        return (2*diff)
    elif diff<0:
        # print(diff)
        return (diff)
    else:
        return "entered number is same"
```

```python
num_check(-10)
```
