```python
size = 1000000

l1 = range(size) # 0, 1, 2, 3
l2 = range(size) # 0, 1, 2, 3

ar1 = np.arange(size)
ar2 = np.arange(size)

start = time.time()
result = [x+y for x, y in zip(l1, l2)]
stop = time.time()

print("Time taken:", (stop - start)*1000, 'ms')
```

```python
size = 1000000

l1 = range(size) # 0, 1, 2, 3
l2 = range(size) # 0, 1, 2, 3

ar1 = np.arange(size)
ar2 = np.arange(size)

start1 = time.time()
result = ar1 + ar2
stop1 = time.time()

print("Time taken:", (stop1 - start1)*1000, 'ms')
```
