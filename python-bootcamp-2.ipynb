## List comprehension

#### traditional approach find the even/ odd number given list 

```python
new_even_list = []
for i in range (1, 11):
        if i%2 ==0:
            new_even_list.append(i)
print(new_even_list)
```

```python
# List comprehension: Elegant way of creating a new list
```

```python
#### Even number

new_even_list = [i for i in range (1, 11) if i % 2 == 0]
new_even_list
```

```python
#### odd number

new_even_list = [i for i in range(0, 11) if i %2 != 0]
new_even_list
```

```python
new_even_list = [i for i in range(0, 10, 2)]
new_even_list
```

```python
list1 = [2,3,4,5,6,7,8]

new_list = [i**2 for i in list1]
new_list
```

### Loops in python:

```python
# 2 types of loops in pyton
1) for loop 
2) while loop

- generate the number
- iterate throught a data structure/sequence
```

```python
# Generate numbers using for loop

for item in range(1, 11):
    print(item)
```

```python
# # Generate numbers using while loop

# 1) initilize the variable
# 2) exit condition
# 3) increment
```

```python
item = 1
while item <11:
    print(item)
    # item = item+1
    item+=1
```
