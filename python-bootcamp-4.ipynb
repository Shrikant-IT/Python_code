```python
## find the squares of even numbers from list of numbers.
```

```python
nums = [3,4,5,1,2,6,8,4,8,9,10,12]

even_sq = []
for x in nums:
    if x % 2 == 0:
        s = x*x
        even_sq.append(s)
        print(s)
```

```python
even_sq

for i in even_sq:
    print(i*i)
```

```python
### Break continiue and pass in python
```

```python
for i in range(0, 11):
    if i == 5:
        # break
        print("Condition matched after that not print anything")
        break
    else:
        # break
        print("condition not matched", i)
```

```python
for i in range(0, 11):
    if i == 5:
        print("Condition matched after that not print anything", i)
        continue
    else:
        print("condition not matched", i)
```
