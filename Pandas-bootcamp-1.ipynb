```python
nums = [5, 4, 10, 5, 0, 7, 50]
```

```python
# pd.Series() -- > convert data into series
```

```python
num_series = pd.Series(nums)
num_series
```

```python
type(num_series)
```

```python
print(num_series.max())
print(num_series.min())
print(num_series.mean())
print(num_series.std())
print(num_series.median())
print(num_series.values)
```

```python
# combine and create two series
```

```python
s1 = pd.Series([10, 20, 30], index=['a', 'b', 'c'], name="s1")
s2 = pd.Series([10, 20, 30], index=['d', 'e', 'f'], name="s2")
```

```python
s1
```

```python
s2
```

```python
# concat = combine multiple series to form dataframe

pd.concat([s1, s2], axis=1) # combine data columnwise
```

```python
T1 = pd.Series([1, 2, 3], index=['a', 'b', 'c'], name="T1")
T2 = pd.Series([4, 5, 6], index=['a', 'b', 'c'], name="T2")
```

```python
new_df = pd.concat([T1, T2], axis=1)  # axis=1 coloum wise 
new_df
```

```python
print(new_df)
```

```python
type(pd.concat([T1, T2], axis=0)) # axis=0 row wise
```

```python
type(new_df)
```
