```python
# take two input from the user (cost price, selling price)
# based on input find profit, loss and no profit no loss
```

```python
cp = int(input("Enter your cost price"))
sp = int(input("Enter your selling price"))

if cp>sp:
    print("you have loss")
elif cp<sp:
    print("you have profit")
else:
    print("No loss no profit")
```

```python
# ask input marks from student and decide grade for him/her
# marks (0-100)
# passing is 35
```

```python
# if elif ladder

marks = int(input("Enter your marks"))

if marks <=0:
    print("Are you mad or what")
elif marks >100:
    print("Be serious and be real...")
elif marks <=35:
    print("you god grade F, just pass")
elif marks >=35 and marks<60:
    print("Grade B")
elif marks >=60 and marks<80:
    print("Grade A")
elif marks >=80 and marks<=100:
    print("Excellent..!!!!")
```
